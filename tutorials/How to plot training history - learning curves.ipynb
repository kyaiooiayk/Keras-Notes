# Import modules

```python
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import numpy
import matplotlib.pyplot as plt
from matplotlib import rcParams
import matplotlib as mpl

rcParams['figure.figsize'] = 17, 8
rcParams['font.size'] = 20
```

# Read-in the dataset

```python
seed = 7
numpy.random.seed(seed)
dataset = numpy.loadtxt("../DATASETS/pima-indians-diabetes.csv", delimiter=",")
# Split into features (X) and outout y
X = dataset[:,0:8]
Y = dataset[:,8]
```

# Create the model

```python
model = Sequential()
model.add(Dense(12, input_dim=8, kernel_initializer = "uniform" , activation= "relu"))
model.add(Dense(8, kernel_initializer = "uniform" , activation= "relu"))
model.add(Dense(1, kernel_initializer = "uniform" , activation= "sigmoid"))

# Compile model
model.compile(loss= "binary_crossentropy" , optimizer= "adam" , metrics=["accuracy"])

# Fit the data
history = model.fit(X, Y, validation_split=0.33, epochs = 500, batch_size=10, verbose=1)
```

```python
# List all the keys
print(history.history.keys())
```

# Plotting learning curves

```python
plt.plot(history.history[ "accuracy" ], lw = 3)
plt.plot(history.history[ "val_accuracy" ], lw = 3)
plt.title( "model accuracy" )
plt.ylabel( "accuracy" )
plt.xlabel( "epoch" )
plt.legend([ "train" ,  "test" ], loc= "upper left" )
plt.show()
```

<div class="alert alert-info">
<font color=black>

- We probably stop the training at around epocj = 300, otherwise we are resking to overfit the model. 

</font>
</div>

```python
plt.plot(history.history[ "loss" ], "k-", lw = 3)
plt.plot(history.history[ "val_loss" ], "r-", lw = 3)
plt.title( "model loss" )
plt.ylabel( "loss" )
plt.xlabel( "epoch" )
plt.legend([ "train" ,  "test" ], loc= "upper left" )
plt.show()
```

<div class="alert alert-info">
<font color=black>

- From the plot of loss, we can see that the model has comparable performance on epoches below 150.
- If these parallel plots start to depart consistently, it might be a sign to **stop training** at an earlier epoch.

</font>
</div>
